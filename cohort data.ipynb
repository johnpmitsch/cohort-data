```python
def week_mean(stupid_chart):
    i = 0
    week_mean_list = []
    while i < 13:
        week_mean_list.append(stupid_chart.transpose()[i:i+4].mean().mean())
        i += 4
    return week_mean_list
```

```python
rails_lecture_week_mean = week_mean(rails_lecture_students)
```

```python
rails_hw_week_mean = week_mean(rails_hw_students)
```

# Rails homework avg mean by week

```python
plt.bar(range(4), rails_hw_week_mean)
plt.xticks(range(0,5), range(1,5))
plt.xlabel("Week")
plt.ylabel("Score Average")
plt.show()
```
